```python
import glob
import matplotlib.pylab as plt
%matplotlib inline
import json
import numpy as np
```

```python
for f in glob.glob('*fgm.json'):    
    with open(f) as inp:
        data = json.loads(inp.read())        
        plt.plot(np.array(data).mean(0)[:,0], np.array(data).mean(0)[:,1], label=f)
        plt.fill_between(np.array(data).mean(0)[:,0], np.array(data).min(0)[:,1], np.array(data).max(0)[:,1], alpha=0.1)
    plt.legend(loc='best')
    plt.rcParams['figure.figsize'] = (16,12 )
```

```python
for f in glob.glob('*struct.json'):    
    with open(f) as inp:        
            data = json.loads(inp.read())  
            #data = np.array(data)
            #        . 
            #       ,  ,   - 
            
            if 'var' in f:
                data = np.array([d for d in data[5:] if len(d)>0]).reshape(5, 11, 2)
            else:
                data = np.array([d for d in data if len(d)>0]).reshape(5, 11, 2)
            data = np.array(data)
            plt.plot(np.array(data).mean(0)[:,0], np.array(data).mean(0)[:,1], label=f)
            plt.fill_between(np.array(data).mean(0)[:,0], np.array(data).min(0)[:,1], np.array(data).max(0)[:,1], alpha=0.1)
        
    plt.legend(loc='best')
    plt.rcParams['figure.figsize'] = (16,12 )
```
